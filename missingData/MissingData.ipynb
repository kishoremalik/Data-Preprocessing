### MissingData1

```python
import numpy as np
import pandas as pd
from collections import Counter
```

```python
class preprocess(object):
    def __init__(self):
        pass
    
    def load_data(self,file):
        data=pd.read_csv(file)
        return data
        
        
    def missing_value(self,data):
        mis_data=data.isnull().sum()
        print("missing data=",mis_data)
        
    def fill_age(self,data):
        fill_data=data.copy()
        age_median=fill_data.Age.median()
        print("median value=",age_median)
        fill_data['Age']=fill_data['Age'].replace(np.NaN,age_median)
        print("missing data=",fill_data.Age.isnull().sum())
        return fill_data
    def cabin_categorical(self,data):
        # changing the dataset to single charachters
        cabin_data=[]
        l=data['Cabin']
        for i in l:
            k=str(i)
            if k !='nan':
                cabin_data.append(k[0])
            else:
                cabin_data.append("NaN")
        
        data['Cabin']=cabin_data
        return data
    def categorical_Map(self,data):
        data_cat=data.copy()
        size_mapping={'A':0,'B':1,'C':2,'D':3,'E':4,'F':5,'G':6,'T':7}
        data_cat['Cabin']=data_cat['Cabin'].map(size_mapping)
        return data_cat
    
    def custom_mode(self,cat_map):
        l=cat_map['Cabin']
        num=[]
        for i in l:
            k=str(i)
            if k in 'nan':
                pass
            else:
                num.append(k)

        f_count=[]
        for n in num:
            f_count.append(float(n))
        v_count=max(k for k,v in Counter(f_count).items() if v>1)
        return v_count

    def categorical_filData(self,data,mode_val):
        data1=data.copy()
        print("categorical fil data")
        c_mode=mode_val
        print("median value=",c_mode)
        data1['Cabin']=data1['Cabin'].replace(np.NaN,c_mode)
        print("missing data=",data1.isnull().sum())
        return data1
        
```
